## Handle Missing Values

```python
calls["Purchase Amount"]
```

```python
calls['Purchase Amount']=calls["Purchase Amount"].fillna(calls["Purchase Amount"].mean(),inplace=True)
```

```python
calls.isnull().sum()
```

```python
calls["Representative"]=calls["Representative"].fillna("unknown",inplace=True)
```
