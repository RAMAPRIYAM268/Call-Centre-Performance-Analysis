## Correlation

```python

calls["Date of Call"].dtype
```

```python
calls["Month"]=calls["Date of Call"].dt.month_name()
```

```python
calls.head()
```

```python
calls.sample(10)
```

```python
# How busy each month
Monthly_calls=calls.groupby("Month")["Call number"].count()
```

```python
Monthly_calls.plot(kind="line",marker="*")
plt.ylabel("No of Calls")
plt.title("Monthly call volume")
plt.show()
```

```python
# which representative most calls
Representative_Calls=calls.groupby("Representative")["Call number"].count()
```

```python
Representative_Calls.plot(kind="bar")
plt.ylabel("Highest calls")
plt.title("HighestCall Representative ")
plt.show()
```

```python
calls["Rating"]=calls["Rating"].round()
```

```python
#Are most customer Satisfaction
calls["Rating"].value_counts().sort_index().plot(kind="pie",autopct="%1.1f%%")
plt.title("Satisfaction distribution")
plt.show()
```

```python
# Call duration vs Purchase amount
plt.scatter(calls["Duration"],calls["purchase"])
plt.xlabel(" call Duration")
plt.ylabel("purchase")
plt.title("Purchase vs Call duration")
plt.show()
```
